***Exercício 1*<h2>Conteúdo importado da APS3 para realização do exercício 1</h2>**

```python
import pandas as pd
df = pd.read_csv('API_NY.GDP.MKTP.KD.ZG_DS2_en_csv_v2_4682262.csv', skiprows=4)
df
```

```python
# lista mercosul
mercosul = ['Argentina', 'Brazil', 'Paraguay', 'Uruguay', 'Venezuela, RB']
dici_media_mercosul = {}
df_mercosul = df[df['Country Name'].isin(mercosul)]

# fazer nova coluna com media
df_mercosul['media'] = df_mercosul.mean(axis=1)

#Pega coluna media
df_mercosul = df_mercosul['media']
print(df_mercosul)

for pais,num in zip(mercosul,range(len(df_mercosul))):
    dici_media_mercosul[pais] = (df_mercosul.iloc[num])

```

```python
#lista de paises da uniao europeia
lista_uniao_europeia = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovak Republic', 'Slovenia', 'Spain', 'Sweden', 'United Kingdom']
dici_media_uniao_europeia = {}
df_uniao_europeia = df[df['Country Name'].isin(lista_uniao_europeia)]

#fazer media dos paises da uniao europeia
df_uniao_europeia['media'] = df_uniao_europeia.mean(axis=1)

#pegar coluna media dos paises da uniao europeia
df_uniao_europeia_media = df_uniao_europeia['media']
df_uniao_europeia_media
#pegar a segunda coluna do dataframe

for pais,num in zip(lista_uniao_europeia,range(len(df_uniao_europeia_media))):
    dici_media_uniao_europeia[pais] = (df_uniao_europeia_media.iloc[num])

```

```python
#gdp medio global
gdp_medio_global = df.mean(axis=1)
#Indice global está na linha 265
gdp_medio_global = gdp_medio_global.iloc[264]
print(gdp_medio_global)

lista_paises_avancados_uniao_europeia = []
lista_paises_emergentes_uniao_europeia = []



lista_paises_avancados_mercosul = []
lista_paises_emergentes_mercosul = []

lista_medias_paises_avancados_uniao_europeia = []
lista_medias_paises_avancados_merco_sul = []

for pais in dici_media_uniao_europeia:
    if dici_media_uniao_europeia[pais] > gdp_medio_global:
        lista_paises_avancados_uniao_europeia.append(pais)
        lista_medias_paises_avancados_uniao_europeia.append(1)
    else:
        lista_medias_paises_avancados_uniao_europeia.append(0)
        lista_paises_emergentes_uniao_europeia.append(pais)

for pais in mercosul:
    if dici_media_mercosul[pais] > gdp_medio_global:
        lista_paises_avancados_mercosul.append(pais)
        lista_medias_paises_avancados_merco_sul.append(1)
    else:
        lista_medias_paises_avancados_merco_sul.append(0)
        lista_paises_emergentes_mercosul.append(pais)
```

```python
prob_uniao_europeia = len(lista_paises_avancados_uniao_europeia)/len(lista_uniao_europeia)
print(f'Probabilidade de um pais da uniao europeia ser avancado: {prob_uniao_europeia*100:.2f} %')

prob_merco_sul = len(lista_paises_avancados_mercosul)/len(mercosul)
print(f'Probabilidade de um pais do mercosul ser avancado: {prob_merco_sul*100:.2f} %')
```

***EXERCÍCIO 2* <h2>Usando bootstrapping, calcule o intervalo de confiança de 90% para P(avançado|bloco) para cada bloco</h2>**

*Intervalo de confiança da União Européia*

```python
from scipy import stats
import numpy as np

amostras_uniao_europeia = lista_medias_paises_avancados_uniao_europeia

def reamostrar_varias_vezes(amostras):
    probs = []
    for i in range(len(amostras)):
        amostras_ = np.random.choice(amostras, size=len(amostras), replace=True)
        probs.append(list(amostras_).count(1)/len(amostras_))
    return np.percentile(probs, 5), np.percentile(probs, 95)

prob_uniao_europeia_5, prob_uniao_europeia_95 = reamostrar_varias_vezes(amostras_uniao_europeia)
```

*Intervalo de confiança do Mercosul*

```python
from scipy import stats
import numpy as np

amostras_mercosul = lista_medias_paises_avancados_merco_sul

def reamostrar_varias_vezes(amostras):
    probs = []
    for i in range(len(amostras)):
        amostras_ = np.random.choice(amostras, size=len(amostras), replace=True)
        probs.append(list(amostras_).count(1)/len(amostras_))
    return np.percentile(probs, 5), np.percentile(probs, 95)

prob_mercosul_5, prob_mercosul_95 = reamostrar_varias_vezes(amostras_mercosul)
```

***Exercício 3*<h2>Figura com barra de erros para demonstrar os intervalos de confiança das médias dos países da União Européia e do Mercosul<h2>**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

plt.figure(figsize=(6,3),dpi=1000)
plt.errorbar('UE', prob_uniao_europeia, yerr=prob_uniao_europeia_95-prob_uniao_europeia_5, fmt='o', color='black', ecolor='blue', elinewidth=3, capsize=10)
plt.bar('UE', prob_uniao_europeia, color='red', width=0.2)
plt.errorbar('mercosul', prob_merco_sul, yerr=prob_mercosul_95-prob_mercosul_5, fmt='o', color='black', ecolor='red', elinewidth=3, capsize=10)
plt.bar('mercosul', prob_merco_sul, color='blue', width=0.2)
plt.title('Probabilidade de ser avançado dado que pertence à União Européia ou ao Mercosul')
plt.ylabel('Probabilidade')
plt.show()
```

Resposta : Há sobreposição entre os gráficos dos intervalos de confiança e as barras de probabilidades em ambos os blocos.
